```python
df=pd.read_csv('data.csv')
print(df)
```

```python
df.fillna(111, inplace=True)     #fillna use to replace the null/ empty cells vth value
print(df)
```

```python
x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)

print(df.to_string())
```

```python
x = df["Calories"].median()

df["Calories"].fillna(x, inplace = True)

print(df.to_string())
```

```python
x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace = True)

print(df.to_string())
```
